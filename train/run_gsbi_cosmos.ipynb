```python
%load_ext autoreload
%autoreload 2

import warnings
warnings.filterwarnings("ignore")

import sys
import os
import pandas as pd
import json
import glob
from IPython.display import clear_output

sys.path.append('../')
from src.external.googlesbi import SearchByImageService
```

```python
input_file = 'G:\\My Drive\\COSMOS\\all_shared_links_of_images.tsv'
output_folder = 'G:\\My Drive\\COSMOS\\test_gsbi\\'
```

```python
links_df = pd.read_csv(input_file, sep='\t', header=None)

processed_files = [f.split('\\')[-1].replace('.json', '') for f in glob.glob(output_folder+'*.json')]

for idx, row in links_df.iterrows():
    print(idx)
    if row[0] in processed_files:
        print('skip')
        continue
    data = SearchByImageService.get_instance().search(
        'https://drive.google.com/uc?id='+row[1],
        limit_page=5
    )
    with open(output_folder+row[0]+'.json', 'w') as out:
        out.write(json.dumps(data, indent=2))
    clear_output(wait=True)
    
```

```python
json_files = glob.glob(output_folder+'*.json')
```

```python
sv = SearchByImageService.get_instance()

for json_file in json_files:
    data = None
    with open(json_file) as jf:
        data = json.loads(jf.read())
    for i in range(len(data)):
        if 'content' not in data[i]:
            continue
        size = sv.find_and_format_size(data[i]['content'])
        if size:
            data[i]['size'] = size
            del data[i]['content']
            continue
        date = sv.find_and_format_datetime(data[i]['content'])
        if date:
            data[i]['date'] = date
            del data[i]['content']
            continue
    if data and len(data) > 0:
        with open(json_file, 'w') as out:
            out.write(json.dumps(data, indent=2))
```
